```python
import json
from transformers import BertTokenizer, BertForTokenClassification
from transformers import pipeline

# Model ve tokenizer'ı yükle
model_name = "dbmdz/bert-base-turkish-cased"  # Türkçe BERT modeli
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForTokenClassification.from_pretrained(model_name)

# Hugging Face pipeline'ını oluştur
ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)

def extract_entities_with_bert(text):
    """
    Türkçe BERT modelini kullanarak metinden entity'leri çıkarır.
    """
    ner_results = ner_pipeline(text)

    # Entities çıkarımı
    entities = []
    for result in ner_results:
        entities.append(result['word'])

    # Token'ları birleştirerek entitiy'leri tekrar birleştiriyoruz
    return list(set(entities))

def process_jsonl(input_file_path, output_file_path):
    """
    JSONL dosyasını okuyup entity'leri çıkararak JSON formatında yazdırır.
    """
    with open(input_file_path, 'r', encoding='utf-8') as infile, \
         open(output_file_path, 'w', encoding='utf-8') as outfile:

        for line in infile:
            data = json.loads(line)
            comment = data.get("comment", "")

            # Türkçe BERT modelini kullanarak entity'leri çıkar
            entities = extract_entities_with_bert(comment)

            # JSON formatında yaz
            result = {"entity_list": entities}
            outfile.write(json.dumps(result, ensure_ascii=False) + '\n')

# Dosya yolları
input_jsonl_file_path = '/content/drive/MyDrive/oflazer/Mert/Sikayetvar_operator/sikayetvar_comments_label.jsonl'  # Girdi dosyasının yolu
output_jsonl_file_path = '/content/drive/MyDrive/oflazer/Mert/Sikayetvar_operator/sikayetvar_comments_entitylabel.jsonl'  # Çıktı dosyasının yolu

# İşlemi başlat
process_jsonl(input_jsonl_file_path, output_jsonl_file_path)
print(f"Entity'ler '{output_jsonl_file_path}' dosyasına kaydedildi.")
```

```python
import json
import re
from transformers import BertTokenizer, BertForTokenClassification
from transformers import pipeline
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk import download

# NLTK kaynaklarını indirin
download('stopwords')
download('punkt')

# Türkçe stopwords ve numara kelimeleri
stop_words = set(stopwords.words('turkish'))

# Sayıları kelimelere dönüştürmek için bir sözlük
number_to_word = {
    '0': 'sıfır', '1': 'bir', '2': 'iki', '3': 'üç', '4': 'dört', '5': 'beş',
    '6': 'altı', '7': 'yedi', '8': 'sekiz', '9': 'dokuz'
}

def preprocess_text(text):
    """
    Metni küçük harfe dönüştürür, noktalama işaretlerini kaldırır,
    stopwords'ü çıkarır ve sayıları kelimelere dönüştürür.
    """
    # Küçük harfe dönüştürme
    text = text.lower()

    # Sayıları kelimelere dönüştürme
    text = ''.join(number_to_word.get(char, char) for char in text)

    # Noktalama işaretlerini kaldırma
    text = re.sub(r'[^\w\s]', '', text)

    # Stopwords çıkarma
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]

    # Token'ları birleştirme
    return ' '.join(tokens)
```

```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForTokenClassification

tokenizer = AutoTokenizer.from_pretrained("akdeniz27/bert-base-turkish-cased-ner")
model = AutoModelForTokenClassification.from_pretrained("akdeniz27/bert-base-turkish-cased-ner")

ner = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="first")

def extract_entities(review):
    entities = ner(review)
    orgs = [entity['word'] for entity in entities if entity['entity_group'] in ['ORG']]
    return orgs

data['entity_list'] = data['comment'].apply(extract_entities)

result = data[['comment', 'entity_list']].to_dict(orient='records')


for entry in result:
    print(entry)
```
