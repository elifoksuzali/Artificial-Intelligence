```python
import re
import pandas as pd
from nltk.util import bigrams, trigrams
from collections import Counter

def preprocess_text(text):
    # Noktalama işaretlerini kaldırma
    text = re.sub(r'[^\w\s]', '', text)

    # Küçük harfe dönüştürme
    text = text.lower()

    # Sayıları ve sayı içeren kelimeleri çıkarma
    text = re.sub(r'\b\d+\b', '', text)  # Sayıları çıkarır
    text = re.sub(r'\b\w*\d\w*\b', '', text)  # Sayı içeren kelimeleri çıkarır

    # Kelime tokenizasyonu
    kelimeler = text.split()

    return kelimeler


# Tüm yorumları ön işleme tabi tut
kelimeler = preprocess_text(tum_yorumlar)

# Bigram ve trigramları oluştur
bigram_list = list(bigrams(kelimeler))
trigram_list = list(trigrams(kelimeler))

# Bigram ve trigram sıklıklarını hesapla
bigram_sikliklari = Counter(bigram_list)
trigram_sikliklari = Counter(trigram_list)

# En sık geçen bigram ve trigramları bul
en_sik_bigramlar = bigram_sikliklari.most_common(10)  # İlk 10 bigram
en_sik_trigramlar = trigram_sikliklari.most_common(10)  # İlk 10 trigram

# Sonuçları göster
print("En sık geçen bigramlar:")
for bigram, frekans in en_sik_bigramlar:
    print(f"{bigram}: {frekans}")

print("\nEn sık geçen trigramlar:")
for trigram, frekans in en_sik_trigramlar:
    print(f"{trigram}: {frekans}")
```
