```python
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk import ne_chunk, pos_tag
from nltk.tree import Tree
import string

# Gerekli NLTK veri setlerini indirme
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('stopwords')

# CSV dosyasını oku
df = pd.read_csv('/content/drive/MyDrive/oflazer/data/deneme.csv')

df
```

```python
# Lemmatizer oluştur
lemmatizer = WordNetLemmatizer()

# Stopwords listesi
stop_words = set(stopwords.words('turkish'))

# Metni ön işleme fonksiyonu

def preprocess_text(text):
    # Küçük harfe çevir, tokenize et, sayıları ve noktalama işaretlerini kaldır
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in string.punctuation]
    tokens = list(set(tokens))  # Tekrarlayan kelimeleri kaldır
    tokens = [token for token in tokens if token not in stop_words]  # Stopwords'leri kaldır
    return tokens
# NER fonksiyonu
    pos_tags = pos_tag(tokens)
    ner_tags = ne_chunk(pos_tags)

    entities = []
    for chunk in ner_tags:
        if isinstance(chunk, Tree):
            entity = " ".join([c[0] for c in chunk])
            entity_type = chunk.label()
            entities.append({'entity': entity, 'type': entity_type})

    return entities

# Yorumları işle
df['entities'] = df['comment'].apply(extract_entities)

# Sonuçları göster
print(df[['comment', 'entities']])
```
