# **veri ön işleme**
### NLTK (Natural Language Toolkit), doğal dil işleme (NLP) görevleri için yaygın olarak kullanılan bir Python kütüphanesidir. NLTK'yi veri ön işleme için kullanmak, dilsel veri temizleme ve dönüştürme işlemlerinde oldukça yararlıdır. NLTK ile veri ön işlemede yapabileceğiniz bazı temel işlemler şunlardır:



*   Metin Tokenizasyonu: Metni kelimelere veya cümlelere bölme.
*   Stop Words: Sık kullanılan ve genellikle bilgi içermeyen kelimelerin (örneğin "ve", "bir", "o") çıkarılması.
*   Kök Bulma (Stemming): Kelimeleri köklerine indirgeme.
*   Lemmatizasyon: Kelimeleri temel formlarına dönüştürme.
*   Etiketleme (POS Tagging): Kelimelerin türlerini belirleme (isim, fiil, sıfat, vb.)
*   Cümle Segmentasyonu: Metni cümlelere bölme.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Gerekli NLTK veri setlerini indir
nltk.download('stopwords')
nltk.download('punkt')

def preprocess_text(text):
    # Noktalama işaretlerini kaldırma
    text = re.sub(r'[^\w\s]', '', text)

    # Küçük harfe dönüştürme
    text = text.lower()

    # Sayıları çıkarma
    text = re.sub(r'\d+', '', text)

    # Kelime tokenizasyonu
    kelimeler = word_tokenize(text)

    # Stop words çıkarma
    stop_words = set(stopwords.words('turkish'))
    kelimeler_filtreli = [kelime for kelime in kelimeler if kelime not in stop_words]

    # Her kelimeden sadece bir tanesini içeren set
    unique_kelimeler = list(set(kelimeler_filtreli))

    # Filtrelenmiş ve tekilleştirilmiş kelimeleri birleştirip metin olarak döndürme
    text = ' '.join(unique_kelimeler)

    return text
```

### preprocess_text fonksiyonu, metindeki noktalama işaretlerini ve sayıları kaldırır, metni küçük harfe dönüştürür, kelimeleri tokenize eder, stop words'leri çıkarır ve her kelimeden sadece bir tanesini içeren bir set oluşturur. Son olarak, kalan kelimeleri birleştirerek temizlenmiş metin olarak döndürür.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Gerekli NLTK veri setlerini indir
nltk.download('stopwords')
nltk.download('punkt')

def preprocess_text(text):
    # Noktalama işaretlerini kaldırma
    text = re.sub(r'[^\w\s]', '', text)

    # Küçük harfe dönüştürme
    text = text.lower()

    # Sayıları çıkarma
    text = re.sub(r'\d+', '', text)

    # Kelime tokenizasyonu
    kelimeler = word_tokenize(text)

    # Stop words çıkarma
    stop_words = set(stopwords.words('turkish'))
    kelimeler_filtreli = [kelime for kelime in kelimeler if kelime not in stop_words]

    # Her kelimeden sadece bir tanesini içeren set
    unique_kelimeler = list(set(kelimeler_filtreli))

    # Filtrelenmiş ve tekilleştirilmiş kelimeleri birleştirip metin olarak döndürme
    text = ' '.join(unique_kelimeler)

    return text
```

```python
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from transformers import pipeline, AutoModelForTokenClassification, AutoTokenizer


# NER modeli ve tokenizer yükleme
model = AutoModelForTokenClassification.from_pretrained("savasy/bert-base-turkish-ner-cased")
tokenizer = AutoTokenizer.from_pretrained("savasy/bert-base-turkish-ner-cased")
ner = pipeline('ner', model=model, tokenizer=tokenizer, aggregation_strategy="first")


# Boş bir DataFrame oluştur
df = pd.DataFrame(columns=["comment", "entity_list"])

for comment in data:
    # Yorumları ön işleme
    processed_comment = preprocess_text(comment)

    # Entity tespiti
    entities = ner(processed_comment)

    # Sadece "ORG" etiketli entitileri filtrele
    org_entities = [entity['word'] for entity in entities if entity['entity_group'] == 'ORG']

    # Yeni satırı DataFrame'e ekle
    new_row = {"comment": processed_comment, "entity_list": org_entities}
    df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
```
