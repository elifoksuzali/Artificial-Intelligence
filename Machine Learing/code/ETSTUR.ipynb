```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd

# WebDriver'ı başlat

# Otel yorumlarının bulunduğu sayfaya git
otel_url = "https://www.etstur.com/Hotel-Abro-Necatibey?&adult_1=2&hotelIndex=8"
driver.get(otel_url)

# Yorumları çekmek için fonksiyon
def scrape_reviews():
    all_user_reviews = []

    # Sayfanın kaynak kodunu al
    page_source = driver.page_source

    # BeautifulSoup ile HTML içeriğini parse et
    soup = BeautifulSoup(page_source, "html.parser")

    # Otel yorumlarının bulunduğu bölümü bulun
    reviews_section = soup.find("div", {"id": "user-reviewsContainer"})

    if reviews_section:
        review_elements = reviews_section.find_all("div", {"class": "userReview-item-wrapper"})

        # Yorumları çıkar
        for review in review_elements:
            comment_element = review.find("li", {"class": "review-list-element userReview-comment-wrapper"})
            #class="userReview-comment-wrapper"
            comment = comment_element.text.strip() if comment_element else "Yorum yok"

            score_element = review.find("span", {"class": "pointMeasure"})
            general_score = score_element.text.strip() if score_element else "Puan yok"

            # Yorum bilgilerini sakla
            all_user_reviews.append({
                'comment': comment,
                'general_score': general_score,
            })

    return all_user_reviews

# "Daha Fazla Göster" düğmesine tıklayarak tüm yorumları yükleme
while True:
    try:
        load_more_button = WebDriverWait(driver, 20).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="user-reviewsContainer"]/div[2]/div/button')))
        load_more_button.click()
        time.sleep(5)  # Buton tıklandıktan sonra yeni içerik yüklendiğinde bekleme süresi//*[@id="user-reviewsContainer"]/div[2]/div/button
    except:
        break  # Buton bulunamadığında veya tıklanamadığında döngüyü sonlandır

# Yorumları çek
all_user_reviews = scrape_reviews()

# Yorumları DataFrame'e dönüştür
hotels = pd.DataFrame(all_user_reviews)

# DataFrame'i yazdır
print(hotels)

# WebDriver'ı kapat
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# WebDriver ayarları (ChromeDriver yolu güncellenmeli)
driver = gs.Chrome(options=options)



# Şehirdeki otel URL'lerini alma (şehir URL listesi)
sehirler = ['Antalya',  'Belek', 'Didim',  'Istanbul','Marmaris','Palandoken',
            'Sapanca','Bodrum', 'Datca','Kapadokya','Mersin','Buyukada','Alanya',
            'Bozcaada', 'Eskisehir','Kas','Uludag',  'Kemer', 'Bursa', 'Fethiye',
            'Amasra', 'Abant', 'Kusadasi', 'Cesme', 'Izmir', 'Ankara', 'Bolu']

sehir_url_listesi = []

for sehir in sehirler:
    url = f"https://www.etstur.com/{sehir}-Otelleri"
    sehir_url_listesi.append(url)

print(f"Ziyaret ediliyor:{sehir_url_listesi}")


# Otel URL'lerini toplamak için şehir sayfasını ziyaret et
hotel_urls = []
for query3 in sehir_url_listesi:
    driver.get(query3)  # Sayfaya istek gönder

    # Sayfanın tamamını yüklemek için "Daha Fazla Göster" düğmesine tıklayın
    while True:
        # Sayfanın en altına kaydır
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(3)  # Kaydırma için kısa bir bekleme süresi

        try:  # "Daha Fazla Göster" düğmesini bulun ve tıklayın
            load_more_button = WebDriverWait(driver, 10).until(
                EC.visibility_of_element_located((By.XPATH, '//*[@id="user-reviewsContainer"]/div[2]/div/button'))

            )
            load_more_button.click()  # Düğmeye tıklayın
            time.sleep(3)  # İçerik yüklendiğinde bekleyin
        except:
            break  # Düğme yoksa veya tıklanamazsa döngüyü sonlandırın

    # Sayfanın kaynak kodunu alın
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")

    # Otel listesini bul
    result_div = soup.find("div", {"class": "result-offers-container", "id": "hotelList"})

    # Otel URL'lerini alın
    if result_div:
        a_tags = result_div.find_all("a", {"class": "hotel-card-link-wrapper"})
        for a_tag in a_tags:
            href_text = a_tag["href"]
            full_url = "https://www.etstur.com/" + href_text  # Tam otel URL'si
            hotel_urls.append(full_url)  # URL'leri saklayın
            print(f"Otel URL'si eklendi: {full_url}")

# Şimdi her bir otel URL'sini ziyaret ederek yorumları çıkar
all_user_reviews = []  # Tüm yorumları saklamak için liste

for otel_url in hotel_urls:
    print(f"Otel sayfası ziyaret ediliyor: {otel_url}")
    driver.get(otel_url)  # Otel sayfasını ziyaret et

    # "Daha Fazla Göster" düğmesine tıklayarak tüm yorumları yükleme
    while True:
        try:
            load_more_button = WebDriverWait(driver, 10).until(
                EC.element_to_be_clickable((By.XPATH, '//*[@id="user-reviewsContainer"]/div[2]/div/button'))
            )
            load_more_button.click()  # Düğmeye tıklayın
            print("Daha Fazla Göster düğmesine tekrar tıklandı...")
            time.sleep(3)  # Yeni içerik yüklendiğinde bekleyin
        except Exception:
            break  # Düğme yoksa veya tıklanamazsa döngüyü sonlandırın

    # Sayfanın kaynak kodunu alın
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")

    # Otel yorumlarının bulunduğu bölümü bulun
    reviews_section = soup.find("div", {"id": "user-reviewsContainer"})

    if reviews_section:
        review_elements = reviews_section.find_all("div", {"class": "userReview-item-wrapper"})

        # Yorumları çıkar
        for review in review_elements:
            comment_element = review.find("li", {"class": "userReview-comment-wrapper"})
            comment = comment_element.text.strip() if comment_element else "Yorum yok"

            score_element = review.find("span", {"class": "pointMeasure"})
            general_score = score_element.text.strip() if score_element else "Puan yok"

            # Yorum bilgilerini sakla
            all_user_reviews.append({
                'comment': comment,
                'general_score': general_score,
            })

# Tüm çıkarılan yorumları yazdırın
for review in all_user_reviews:
    print(f"Yorum: {review['comment']}")
    print(f"Puan: {review['general_score']}")
    print("------")

# WebDriver'ı kapat
driver.quit()
```
