## [1. Turkish Movie Sentiment Analysis Dataset](https://www.kaggle.com/datasets/mustfkeskin/turkish-movie-sentiment-analysis-dataset)
### About Dataset
Context
This dataset contains turkish movie reviews.
Content
Comment --> String --> Contatins turkish movie review


Film_name --> String --> Film name in turkish


Point --> Float --> [0-5] floating point

Öneri:
Film yorumları puan üzerinden etiketlenmiş puan aralığı 3 e kadar olan negative, 3 ten 4’e kadar olan ise nötr, 4 ve üzeri positive diye etiketlenmesi..


```python
import pandas as pd

data = pd.read_csv('/content/drive/MyDrive/oflazer/data/TurkishMovieSentimentAnalysisDataset/turkish_movie_sentiment_dataset.csv')
data
```

```python
# Veriyi DataFrame'e dönüştürme
df = pd.DataFrame(data)
df
```

```python
# Puanları etiketleme
df['point'] = df['point'].astype(str).str.replace(',','.').astype(float)

def label_sentiment(point):
    if point <= 2.5:
        return 'olumsuz'
    elif point < 4:
        return 'notr'
    else:
        return 'olumlu'
df['sentiment'] = df['point'].apply(label_sentiment)

# Yalnızca Comment ve Sentiment sütunlarını seçme
new_df = df[['comment', 'sentiment']]
```

```python
# Yeni veri setini gösterme
print(new_df)
```

```python
new_df.to_csv('/content/drive/MyDrive/oflazer/data/oflazer4tddi.csv', index=False, encoding='utf-8')
```

### **[2.Turkish Sentiment Analysis Data - beyazperde.com](https://www.kaggle.com/datasets/ozcan15/turkish-sentiment-analysis-data-beyazperdecom?resource=download&select=train.csv)**
Veri Seti Hakkında

 Burada, duygu analizi modellerinizi eğitmek için beyazperde.com kullanıcı yorumları verilerini bulabilirsiniz.

Orijinal veri seti (Demirtas ve Pechenizkiy, 2013) 1 ile 5 arasında değişen sınıflara sahip 10662 yoruma sahiptir. Yorumların yıldız derecelendirmeleri 2 sınıfa ayrılmıştır. 4 veya 5 yıldızlı yorumlar olumlu yorumlar olarak kabul edilirken; 1 veya 2 yıldızlı yorumlar olumsuz olarak kabul edilir. 3 yıldızlı yorumlar hariç tutulur.

Son veriler mükemmel bir şekilde dengelenmiştir. Eğitim verilerinde 3998 pozitif etiketli ve 3998 negatif etiketli yorum vardır. Test verilerine gelince, her sınıf için 1333 yorum vardır.

İçerik
Verinin yapısı oldukça basittir. Kullanıcı yorumları ve ilgili sınıfları olan 2 sütun vardır. 0 negatif sınıfı, 1 ise pozitif sınıfı belirtir.

NOT !! biz bu veri setinde 0 ve 1 etiketlerini olumlu ve olumsuz diye etiketleyeceğiz nötr etiketi onlar kullanmadığı için datamızı sadece oumlu ve olumsuz etiketler için beslemiş olacağız

NOT2 !! test ve train data setlerini birleştirip düzenleme işlemi yapılacaktır.

```python
train_data = pd.read_csv("/content/drive/MyDrive/oflazer/data/TurkishSentimentAnalysisData-beyazperde.com/train.csv", encoding='ISO-8859-9')
test_data = pd.read_csv("/content/drive/MyDrive/oflazer/data/TurkishSentimentAnalysisData-beyazperde.com/test.csv", encoding='ISO-8859-9')

combined_df = pd.concat([train_data, test_data], ignore_index=True)
```

```python
df = pd.DataFrame(combined_df)
df
```

```python
# Etiketleri dönüştürme
def convert_labels(label):
    if label == 0:
        return 'olumsuz'
    elif label == 1:
        return 'olumlu'

combined_df['sentiment'] = combined_df['Label'].apply(convert_labels)


new_df = combined_df[['comment', 'sentiment']]
```

```python
new_df
```

```python
# Var olan CSV dosyasına yeni veriyi yazdırma  (var olan dosya üzerine yazmak için mode='a' kullanılır)
new_df.to_csv('/content/drive/MyDrive/oflazer/data/oflazer4tddi.csv', index=False, encoding='utf-8', mode='a', header=False)
```

#[3. duygu-analizi-icin-urun-yorumlari](https://www.kaggle.com/datasets/burhanbilenn/duygu-analizi-icin-urun-yorumlari)

Veriler, Türkçe olarak, çeşitli elektronik mağazalardan toplanmıştır. Bu yüzden tekrar eden değerler olabilir. Modellerinizde daha başarılı sonuçlar elde etmek için verileri ön işleme aşamasından geçirmeniz sizin yararınıza olacaktır.

Duygu ayrımı, "Olumlu", "Olumsuz" ve "Tarafsız" olmak üzere üçe ayrılmıştır.

Bazen, görüşlerle uyuşmayan duygular görebilirsiniz, bunlar tamamen yorumu yapan kişiden kaynaklıdır.

Veri seti .csv uzantılı olup, utf-16 olarak kodlanmıştır. Projelerinizde, veriyi okurken bunları dikkate almanız gerekecektir.

(Turkish reviews for sentiment analysis.)

```python
import chardet
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/oflazer/data/DuyguAnaliziVeriSeti/magaza_yorumlari_duygu_analizi.csv', encoding="utf-16")
print(df)
```

```python
# Sütun adlarını değiştirme
data = df.rename(columns={'Görüş': 'comment', 'Durum': 'sentiment'})
data
```

```python
data['sentiment'] = data['sentiment'].replace({'Olumlu': 'olumlu', 'Olumsuz': 'olumsuz', 'Tarafsız': 'notr'})
data
```

```python
data = data.dropna(subset=['comment'])  # 'comment' sütunundaki boş değerleri kaldırma
data = data[data['comment'].apply(lambda x: isinstance(x, str))]  # 'comment' sütunundaki string olmayan değerleri kaldırma
data
```

```python
# Var olan CSV dosyasına yeni veriyi yazdırma  (var olan dosya üzerine yazmak için mode='a' kullanılır)
data.to_csv('/content/drive/MyDrive/oflazer/data/oflazer4tddi.csv', index=False, encoding='utf-8', mode='a', header=False)
```
