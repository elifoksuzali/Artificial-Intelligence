```python
class Dugum():
    def __init__(self, nameValue, tekrar, parentNode):
        self.name = nameValue
        self.count = tekrar
        self.nodeLink = None
        self.parent = parentNode #Çocukları sürekli güncelleyeceğiz
        self.children = {} 
#sayacı arttırma yapacağız (count) her o düğümde tekrar geçiş yapıyorsak artmasını sağlamamız gerekiyor
    def Gezinme(self, tekrar):
        self.count += tekrar
#Ağaca yerleştirme yapıldıktan sonra ağacı hatalı mı değil mi diye kontrol etmek için göstermemiz gerekiyor     
    def disp(self, a=1):
        print ('  '*a, self.name, ' ', self.count)
        for child in self.children.values():
            child.disp(a+1)
```

```python
def createTree(veriSeti, minSup):
    headerTable = {}
    
    for trans in veriSeti: ##ilk geçişten sonra kaç kere daha geçtiği sıklığını say
        for item in trans:
            ##if nan değer varsa alma onu nasıl yapacağız
            headerTable[item] = headerTable.get(item, 0) + veriSeti[trans]
    for k in list(headerTable):  #
        if headerTable[k] < minSup: 
            del(headerTable[k])
    freqItemSet = set(headerTable.keys())
    print('freqItemSet: ',freqItemSet)
    if len(freqItemSet) == 0: return None, None  #Bütün  öğeler eğer minimum destekten büyük değilse bu bloktan -->çık 
    for k in headerTable:
        #Düğümlerin birbiri ile olan bağlantılarını kullanmak için headerTable yeniden güncelle
        headerTable[k] = [headerTable[k], None]
    #print 'headerTable: ',headerTable
    retTree = Dugum('Null Set', 1, None) #Ağacı oluştur
    for tranSet, count in veriSeti.items():  #2. kez veri setinden dolaş
        localD = {}
        for item in tranSet:  #Öğelerini sırala
            if item in freqItemSet:
                localD[item] = headerTable[item][0]
        if len(localD) > 0:
            orderedItems = [v[0] for v in sorted(localD.items(), key=lambda p: p[1], reverse=True)]
            #Ağaca sıraladığımız sık öğe kümelerine göre düğümleri yerletşir
            updateTree(orderedItems, retTree, headerTable, count)
    return retTree, headerTable #header ve ağacı her seferinde güncelle
```

```python
def updateTree(items, inTree, headerTable, count):
    if items[0] in inTree.children:#Ağacın içinde o düğümün öyle bir çocuğu var mı yok mu kontrol yap
        inTree.children[items[0]].Gezinme(count) #her var olan çocuğu görünce tekrar değerini arttır
    else:   #eğer öyle bir çocuk oluşmadıysa onu ekle
        inTree.children[items[0]] = Dugum(items[0], count, inTree)
        if headerTable[items[0]][1] == None: #başlık tablosunu güncelle
            headerTable[items[0]][1] = inTree.children[items[0]]
        else:
            updateHeader(headerTable[items[0]][1], inTree.children[items[0]])
    if len(items) > 1:#Diğer geriye kalan öğeler için fonksiyonu çağır işleme devam et
        updateTree(items[1::], inTree.children[items[0]], headerTable, count)
```

```python
def updateHeader(nodeToTest, targetNode):   
    while (nodeToTest.nodeLink != None):    #Başlık tablosunu bir kez işlem yaptıkan sonra tekrar o elemana dönmüyoruz
        nodeToTest = nodeToTest.nodeLink
    nodeToTest.nodeLink = targetNode
```
