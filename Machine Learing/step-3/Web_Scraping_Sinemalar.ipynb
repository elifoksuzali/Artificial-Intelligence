Öncelikle siteye erişebilmek için `getAndParseURL` fonksiyonumuzu tanımlayalım.

````python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Safari/537.36."})
    soup = bts(result.text, "html.parser")
    return soup
````

Artık DataFrame'i oluşturma zamanı :)

````python
features = []
for movie in movies:
    html = getAndParseURL(movie)
    try:
        name = html.find("div",{"class":"title"}).text.strip()
    except:
        name = np.nan
    try:
        rating = html.find("span",{"class":"rating-value"}).text.replace(",",".")
    except:
        rating = np.nan
    try:
        movie_type = html.find("span",{"itemprop":"genre"}).text
    except:
        movie_type = np.nan
    try:
        director = html.find("div",{"class" : "content-info"}).find(text=re.compile("Yönetmen:")).findNext().text.strip()
    except:
        director = np.nan
    try:
        rate_count = html.find("span",{"class" : "rate-count"}).text
    except:
        rate_count = np.nan
    try:
        watch_count = html.find("div",{"class" : "left-column"}).find(text=re.compile("Filmi İzleyenler:")).findNext().text
    except:
        watch_count = np.nan
    try:
        fav = html.find("div",{"class" : "left-column"}).find(text=re.compile("Favori:")).findNext().text
    except:
        fav = np.nan
    
        
    features.append([name, rating, movie_type, director, rate_count, watch_count, fav])
    time.sleep(1)
    
df_columns = ["Name","Avg Rating","Movie Type","Director","Rate Count","Watch Count","Fav"]

df = pd.DataFrame.from_records(features, columns=df_columns)
````
