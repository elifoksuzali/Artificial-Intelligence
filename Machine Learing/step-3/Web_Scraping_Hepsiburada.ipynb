```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
import re
import numpy as np
import time
```

```python
def getAndParseURL(url):
    result = requests.get(url, headers={"User-Agent":"Mozilla/5.0"}) # Safari/537.36. Chrome/103.0.0.0
    soup = bts(result.text, "html.parser")
    return soup
```

```python
pages = ["https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98"]
for page in range(2,51):
    pages.append("https://www.hepsiburada.com/laptop-notebook-dizustu-bilgisayarlar-c-98?sayfa="+str(page))

pages
```

```python
products = []
for page in pages:
    html = getAndParseURL(page)
    for product in html.findAll("li",{"class":"productListContent-zAP0Y5msy8OHn5z7T_K_"}):
        products.append("https://www.hepsiburada.com" + product.a.get("href"))

products
```

```python
len(products)
```

```python
links = []
for element in products:
    if element.startswith("https://www.hepsiburada.com/"):
        links.append(element)
```

```python
links
```

```python
len(links)
```

```python
features = []
for link in links:
    html = getAndParseURL(link)
    try:
        brand = html.find("span",{"class":"brand-name"}).text.strip()
    except:
        brand = np.nan
    try:
        rating = html.find("span",{"class":"rating-star"}).text.strip().replace(",",".")
    except:
        rating = np.nan
    try:
        comment = html.find("div",{"id":"comments-container"}).span.text
    except:
        comment = np.nan
    try:
        seller_rating = html.find("span",{"class":"merchant-rating lv-5"}).text.strip()
    except:
        seller_rating = np.nan
    try:
        notebook_type = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ürün Modeli")).findNext().text.strip()
    except:
        notebook_type = np.nan
    try:
        memory_speed = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Bellek Hızı")).findNext().text.strip()
    except:
        memory_speed = np.nan
    try:
        bluetooth = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Bluetooth Özelliği")).findNext().text.strip()
    except:
        bluetooth = np.nan
    try:
        weight = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Cihaz Ağırlığı")).findNext().text.strip()
    except:
        weight = np.nan
    try:
        touch_screen = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Dokunmatik Ekran")).findNext().text.strip()
    except:
        touch_screen = np.nan
    try:
        screen_size = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ekran Boyutu")).findNext().text.strip()
    except:
        screen_size = np.nan
    try:
        gpu_type = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ekran Kartı Tipi")).findNext().text.strip()
    except:
        gpu_type = np.nan
    try:
        gpu = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ekran Kartı")).findNext().text.strip()
    except:
        gpu = np.nan
    try:
        cpu_gen = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("İşlemci Nesli")).findNext().text.strip()
    except:
        cpu_gen = np.nan
    try:
        cpu_type = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("İşlemci Tipi")).findNext().text.strip()
    except:
        cpu_type = np.nan
    try:
        os = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("İşletim Sistemi")).findNext().text.strip()
    except:
        os = np.nan
    try:
        max_cpu_speed = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Maksimum İşlemci Hızı")).findNext().text.strip().replace(",",".")
    except:
        max_cpu_speed = np.nan
    try:
        screen_resolution = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Max Ekran Çözünürlüğü")).findNext().text.strip()
    except:
        screen_resolution = np.nan
    try:
        finger_print = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Parmak İzi Okuyucu")).findNext().text.strip()
    except:
        finger_print = np.nan
    try:
        ram = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ram")).findNext().text.strip()
    except:
        ram = np.nan
    try:
        ram_type = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Ram Tipi")).findNext().text.strip()
    except:
        ram_type = np.nan
    try:
        hdd = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Harddisk Kapasitesi")).findNext().text.strip()
    except:
        hdd = np.nan
    try:
        ssd = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("SSD Kapasitesi")).findNext().text.strip()
    except:
        ssd = np.nan
    try:
        charge_time = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Şarjlı Kullanım Süresi")).findNext().text.strip()
    except:
        charge_time = np.nan
    try:
        cpu_speed = html.find("table",{"class":"data-list tech-spec"}).find(text=re.compile("Temel İşlemci Hızı")).findNext().text.strip().replace(",",".")
    except:
        cpu_speed = np.nan
    try:
        price = html.find("span",{"data-bind":"markupText:'currentPriceBeforePoint'"}).text.replace(".","") + "." + html.find("span",{"data-bind":"markupText:'currentPriceAfterPoint'"}).text
    except:
        price = np.nan

    features.append([brand, rating, comment, seller_rating, notebook_type, memory_speed, bluetooth, weight,
                     touch_screen, screen_size, gpu_type, gpu, cpu_gen, cpu_type, os, max_cpu_speed,
                     screen_resolution, finger_print, ram, ram_type, hdd, ssd, charge_time, cpu_speed, price])
    time.sleep(1)

df_columns = ["Brand","Avg Rating","Total Comment","Seller Rating","Notebook Type","Memory Speed","Bluetooth","Weight",
             "Touch Screen","Screen Size","GPU Type","GPU","CPU Generation","CPU Type","OS","Max CPU Speed",
             "Screen Resolution","Finger Print","RAM","RAM Type","HDD","SSD","Charging Time","CPU Speed","Price"]

df = pd.DataFrame.from_records(features, columns=df_columns)
```
