```python
import requests
from bs4 import BeautifulSoup
import csv

# Otel web sitesinin URL'sini girin
url = "https://www.mngturizm.com/elit-turlar"

# Web sitesinin içeriğini al
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Kazınacak verilerin olduğu HTML elemanlarını bulun
# Bu, web sitesinin yapısına bağlı olarak değişir. Örneğin:
tours = soup.find_all('div', class_='tour')  # Örnek bir sınıf adı

# Verileri saklamak için bir liste oluşturun
data = []

for tour in tours:
    # Her turun kodunu, adını ve içeriğini kazıyın
    tur_kodu = tour.find('span', class_='tur-kod').text.strip()  # Örnek bir sınıf adı
    tur_adi = tour.find('h3', class_='title').text.strip()  # Örnek bir sınıf adı
    icerik = tour.find('div', class_='t-price').text.strip()  # Örnek bir sınıf adı
    data.append([tur_kodu, tur_adi, icerik])

# Verileri CSV dosyasına yazın
with open('turlar.csv', mode='w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Tur Kodu", "Tur Adı", "İçerik"])  # Başlık satırı
    writer.writerows(data)

print("Veriler başarıyla 'turlar.csv' dosyasına yazıldı.")
```
