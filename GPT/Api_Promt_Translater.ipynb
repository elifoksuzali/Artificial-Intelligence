```python
import os
import openai
import pandas as pd
from langchain.chains import LLMChain
from langchain import PromptTemplate
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage

# OpenAI API anahtarını ayarlayın
open_api_key = "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"  # API anahtarınızı buraya koyun
os.environ["OPENAI_API_KEY"] = open_api_key

# CSV dosyasını yükleyin
file_path = "0_pos_1_neg_ispanyol.csv"  # CSV dosyasının doğru yolunu belirtin
data = pd.read_csv(file_path)

# ChatOpenAI modelini ayarlayın
llm = ChatOpenAI(model_name='gpt-3.5-turbo')

# Özetleme ve çeviri için şablon
generic_template = '''
Write a summary of the following text:
Text: `{data}`
Translate the precise summary to {language}.
'''

# Prompt şablonunu oluşturun
prompt = PromptTemplate(
    input_variables=['data', 'language'],
    template=generic_template
)

# LLMChain'i ayarlayın
llm_chain = LLMChain(llm=llm, prompt=prompt)

# Çeviri ve özetleme için bir fonksiyon oluşturun
def summarize_and_translate(text, target_language):
    result = llm_chain.run({'data': text, 'language': target_language})  # Özetleme ve çeviri
    return result

# "speech" sütununu İngilizce'ye özetleyip çevirmek ve yeni bir "translated_summary" sütunu olarak eklemek için
data["translated_summary"] = data["labels"].apply(lambda x: summarize_and_translate(x, "English"))

# Çevrilmiş ve özetlenmiş sonuçları görüntüleyin
print("Çevrilmiş ve Özetlenmiş Metinler:")
print(data[["speech", "translated_summary"]])  # Orijinal ve yeni sütunları gösterin
```
