```python
import os
from tensorflow.keras.preprocessing import image
import numpy as np
from keras.models import load_model
import shutil

dir_path = 'Output_image_19Bin/İç Mekan'  # Test veri setinin bulunduğu klasör
model = load_model('models/ResNet_10_epoch_10_patience_224_size_val_loss.h5')  # Eğitilmiş modeliniz
output_folder = 'output/'  # Sınıflandırılan resimlerin taşınacağı klasör
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

classes = {
    0: 'Dıştan Görünüm',
    1: 'Odalar',
    2: 'Olanaklar',
    3: 'İç Mekan'
}

images_taken = []  # Alınan resimlerin listesi

for i in os.listdir(dir_path):
    img = image.load_img(os.path.join(dir_path, i), target_size=(224, 224))
    
    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    predictions = np.argmax(model.predict(X), axis=1)
    predicted_class = classes[predictions[0]]

    # Resmi ilgili klasöre taşı
    class_folder = os.path.join(output_folder, predicted_class)
    os.makedirs(class_folder, exist_ok=True)
    
    file_name = os.path.basename(i)
    new_file_path = os.path.join(class_folder, file_name)
    shutil.copy(os.path.join(dir_path, i), new_file_path)
    images_taken.append(i)  # Alınan resimlerin listesine ekle

    print(f"Resim: {i}, Tahmin: {predicted_class}, Taşındı: {new_file_path}")

# Alınan resimlerin klasöründen sil
for image_name in images_taken:
    os.remove(os.path.join(dir_path, image_name))
```
