```python
import warnings
warnings.filterwarnings("ignore")
from keras.models import load_model
from tensorflow.keras.preprocessing import image
import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Modeli yükle
model = load_model('Models/ResNet_patience_10_epoch_10_patience_224_size_val_loss.h5')

# Sınıf etiketlerini içeren bir sözlük
class_labels = {"Dıştan Görünüm": "1", "Odalar": "2", "Olanaklar": "3", "Yemek-Restoran": "4"}

dir_path = 'test'

# Sınıf etiketlerini içeren bir DataFrame oluştur
df = pd.DataFrame({'names': list(class_labels.keys()), 'index': list(class_labels.values())})

for i in os.listdir(dir_path):
    img = image.load_img(os.path.join(dir_path, i), target_size=(224, 224))
    plt.imshow(img)
    plt.show()

    X = image.img_to_array(img)
    X = np.expand_dims(X, axis=0)
    images = np.vstack([X])
    predictions = np.argmax(model.predict(images), axis=1)
    
    # Tahmin edilen sınıf etiketini sınıf numarası ile ekrana yazdır
    print("\n",predictions[0] + 1)  # Sınıflar 1'den başladığı için 1 eklenir
```
