```python
import numpy as np
```

```python
x1 = np.array([[1],
              [-2],
              [0],
              [-1]])

x2 = np.array([[0],
               [1.5],
               [-0.5],
               [-1]])

x3 = np.array([[-1],
               [1],
               [0.5],
               [-1]])

x4 = np.array([[-1],
               [1],
               [0.5],
               [-1]])
```

```python
x5 = np.array([
    [-2,4,-1],
    [4,1,-1],
    [1, 6, -1],
    [2, 4, -1],
    [6, 2, -1]])
```

```python
w1 = np.array([1,-1,0,0.5])
```

```python
d1 = -1
d2 = -1
d3 = 1
```

```python
c = 0.1
```

```python
def activation_fn(x):
    if x >= 0:
        return 1
    elif x < 0:
        return -1 
```

```python
def Perceptron(w,x):
    sonuc = np.dot(w,x)
    girdi = 5
    while girdi > 0:
        print("1.Aşama")
        W = w1
        if(d1 == activation_fn(sonuc)):
            print("w1: ",W)         
            
        elif(d1 != activation_fn(sonuc)):
            cikti = W + c*(d1 - activation_fn(sonuc))*x1
            w2 = np.array([cikti[0,0],
                           cikti[1,1],
                           cikti[2,2],
                           cikti[3,3]
                         ])
            W = w2
            print("w2: ",W)
            
        girdi = girdi -1 
    
        print("2.Aşama")
        sonuc = np.dot(W,x2)
        if(d2 == activation_fn(sonuc)):
            W = w2
            print("w2: ",W)
            
        elif(d2 != activation_fn(sonuc)):
            cikti = W + c*(d2 - activation_fn(sonuc))*x2
            w3 = np.array([cikti[0,0],
                           cikti[1,1],
                           cikti[2,2],
                           cikti[3,3]
                         ])
            W = w3
            print("w3: ",W)
            
        girdi = girdi -1       
                        
        print("3.Aşama")
        sonuc = np.dot(W,x3)
        if(d3 == activation_fn(sonuc)):
            W = w2
            print("w2: ",W)
        elif(d3 != activation_fn(sonuc)):
            cikti = W + c*(d3 - activation_fn(sonuc))*x3
            w3 = np.array([cikti[0,0],
                           cikti[1,1],
                           cikti[2,2],
                           cikti[3,3]
                         ])
            W = w3
            print("w3: ",W)
            
        girdi = girdi -1
    
```

```python
Perceptron(w1,x1)
```

```python
#Farklı w1 giriş ile denenmesi:
w1 = np.array([-1,-1,2,1])
Perceptron(w1,x1)
```

```python
#Farklı w1 giriş ile denenmesi:
w1 = np.array([0.5,-1,1,0])
Perceptron(w1,x1)
```

```python
#Farklı c değeri ile denenmesi:
c = 1
Perceptron(w1,x1)
```
